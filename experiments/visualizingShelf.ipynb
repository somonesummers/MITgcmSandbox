```python
from MITgcmutils import mds
import numpy as np
import matplotlib.pyplot as plt
import os

print(os.getcwd())

dir = 'S_500/results'

yslice = 5

XC = mds.rdmds('%s/XC' % dir)
YC = mds.rdmds('%s/YC' % dir)
ZC = mds.rdmds('%s/RC' % dir)

ZC = ZC.squeeze()
XC = XC[yslice,:].squeeze()
print("X", np.shape(XC))
print("Y", np.shape(YC))
print("Z", np.shape(ZC))

endStep = 10
```

```python
T = mds.rdmds('%s/T' % dir, endStep)
T = T[:,yslice,:].squeeze()
print("T: ", np.min(T) , " " , np.max(T), np.shape(T))

W = mds.rdmds('%s/W' % dir, endStep)
W = W[:,yslice,:].squeeze()
print("W: ", np.min(W) , " " , np.max(W), np.shape(W))

U = mds.rdmds('%s/U' % dir, endStep)
U = U[:,yslice,:].squeeze()
print("U: ", np.min(U) , " " , np.max(U), np.shape(U))

S = mds.rdmds('%s/S' % dir, endStep)
S = S[:,yslice,:].squeeze()
print("S: ", np.min(S) , " " , np.max(S), np.shape(S))
```

```python

plt.figure(1)
plt.contourf(XC, ZC, T, np.linspace(-1.8, .4, 12), cmap='hot_r')
plt.colorbar()
plt.title('Temperature')
# plt.savefig("temp.png")
plt.show()
```

```python
plt.figure(2)
plt.contourf(XC, ZC, U, np.linspace(-0.02, 0.02, 32), cmap='bwr')
plt.colorbar()
plt.title('U velocity')
# plt.savefig("U.png")
plt.show()
```

```python
plt.figure(2)
plt.contourf(XC, ZC, W, np.linspace(-0.02, 0.02, 12), cmap='bwr')
plt.colorbar()
plt.title('W velocity')
# plt.savefig("W.png")
plt.show()
```

```python

plt.figure(4)
plt.contourf(XC, ZC, S, np.linspace(33, 35, 64), cmap='bwr')
plt.colorbar()
plt.title('Salt')
plt.show()
```
